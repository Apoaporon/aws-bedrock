```python
import fitz # PyMuPDFのメインモジュール
# チャンキングに利用できそうなライブラリをインストール
from langchain.text_splitter import NLTKTextSplitter, SpacyTextSplitter
```

```python
# pdfを開く
pdf_documents = fitz.open("./knowlege_file/SBG_report.pdf")
# pdfのページをカウントできる
pages = pdf_documents.page_count
print(pages)
```

```python
# テキストと画像を抽出(表情報はテキストに含まれるので、表形式でデータを入れたいのであればちょっと応用が必要かも)
text_items = []
image_items = []

for page_no in range(pages):
    page = pdf_documents[page_no]
    # ページごとのテキストの抽出
    text = page.get_text()
    text_items.append(text)

# pdfを閉じる
pdf_documents.close()


    
```

```python
import spacy
import itertools
import requests
import numpy as np
import spacy.cli
import spacy.cli.download
# all_text = "\n".join(text_items)
spacy.cli.download("ja_core_news_sm")
nlp = spacy.load("ja_core_news_sm")
# 文 -> sentence
sentence_list = [nlp(text).sents for text in text_items]
# 全ページフラットにして結合
sentences = list(itertools.chain.from_iterable(sentence_list))
# 空文字のみのsentenceを取り除く
def remove_empty_strings(lst):
    return [s for s in lst if s.text.strip()]

sentences = remove_empty_strings(sentences)
sentences = [sent.text for sent in sentences]
print(sentences)
```

```python
print(len(sentences))
print(type(sentences))
all_text = "\n".join(text_items)
all_text2 = "\n".join(sentences)
print(all_text2)
```

```python
# NLTK手法で分割
import nltk
# トークナイザーをダウンロード
nltk.download("punkt")
# オブジェクト生成
text_splitter = NLTKTextSplitter(chunk_size=100, chunk_overlap=0)
# 分割
docs = text_splitter.split_text(all_text2)
print(docs)
```

```python
# 分割後の結果を見てみる
print(len(docs))
print(docs[0])
print("##################")
print(docs[1])
```
